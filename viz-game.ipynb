# Data Visulization
- Data : Top Games on Google Play Store

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
android_games = pd.read_csv("./android-games.csv").dropna()
android_games.head()
```

- Compaire between average rating and paid and unpaid games.

```python
g = sns.catplot(
    data=android_games, kind="bar",
    x="paid", y="average rating"
)
g.despine(left=True)
g.set_axis_labels("paid", "average rating")
```

Paid games have average rating larger than unpaid games.

- the relation between growthing in 30 days and growthing in 60 days.

```python
g = sns.catplot(
    data=android_games.tail(20), kind="bar",
    x="growth (30 days)", y="growth (60 days)"
)
g.despine(left=True)
g.set_axis_labels("growth (30 days)", "growth (60 days)")
```

```python
g = sns.catplot(
    data=android_games.head(20), kind="bar",
    x="growth (30 days)", y="growth (60 days)"
)
g.despine(left=True)
g.set_axis_labels("growth (30 days)", "growth (60 days)")
```

- most likley growthing good or bad in 30 days will be the same growthing in 60 days.
