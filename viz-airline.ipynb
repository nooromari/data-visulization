# Data Visulization
- Data : Airline Passenger Satisfaction

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
airline_data = pd.read_csv("./airline.csv").dropna()
airline_data.head()
```

- What is largest age between classes and genders?

```python
g = sns.catplot(
    data=airline_data, kind="bar",
    x="Gender", y="Age", hue='Class'
)
g.despine(left=True)
g.set_axis_labels("", "Age (Y)")
```

the business class has the largest age, and it is close between two genders.

- What are the highest Arrival Delay class?

```python
g = sns.catplot(
    data=airline_data, kind="bar",
    x="Class", y="Arrival Delay in Minutes"
)
g.despine(left=True)
g.set_axis_labels("Class", "Arrival (min)")
```

 The highest Arrival Delay in Minutes are from Eco class.
